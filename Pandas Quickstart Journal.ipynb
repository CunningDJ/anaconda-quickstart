```python
# importing the pandas library.  This first and foremost uses DataFrames, which are objects that 
#  import and manage tabular data (excel, database, etc.)
# The library is called pandas, but we're using pd as a shorter alias (easier to type)
import pandas as pd
```

```python
# Creating a DataFrame object
df = None
# Creating a file object out of the csv file in this directory so it can be read
with open('us-500.csv', 'r') as f:
    # Reading the CSV file in this directory and parsing it into a DataFrame object
    #  using pandas library's read_csv convenience method.  This reads comma-separated data
    #  pandas has other read_* methods (excel, sql, etc.) for other common tabular file formats
    df = pd.read_csv(f)
```

```python
# Run this cell to see the newly-created dataframe, visualized
df
```

```python
# As you can see after running the cell above, it has automatically grabbed the data 
#  from the header row and used it as the names of each column
# Let's grab one of the columns (first_name) to look at
df.first_name
```

```python
# Now let's iterate through all the first names, and print them in uppercase
for name in df.first_name:
    print(name.upper())
```

```python
# Now let's take all the first and last names, and print them together
for index, row in df.iterrows():
    print(row.first_name + " " + row.last_name)
```

```python
# Now, lets do the same, but writing it to a file in this directory, named
#  full-names-file.txt
FULL_NAMES_FILENAME = "full-names-file.txt"
# Again, we're creating a file object first, but this time to write to. 
# Note that the 'w' in the open function means write mode, just as the 'r' in the cell above
#  meant read mode
# Also note that we're adding a newline to the string each time we write to 
#  the file, so it doesn't write all the names on one line.  This acts as a return carriage.
with open(FULL_NAMES_FILENAME, 'w') as f:
    for index, row in df.iterrows():
        f.write(row.first_name + " " + row.last_name + "\n")
```

Now, navigate to the directory you're currently in with your file explorer, and open the file named full-names-file.txt.  You should see a list of these full names printed there!

```python
# Here we'll read the new file for good measure, printing out each newline-stripped line one by one
with open('full-names-file.txt', 'r') as f:
    for line in f:
        print(line.strip())
```

```python
# Now we're going to print out more complex strings/sentences using formatted string
#  (begins with f'... or f"...) for simplicity
for index, row in df.iterrows():
    print(f'{row.first_name} {row.last_name} lives in {row.city}, {row.state} ({row.zip})')
```
